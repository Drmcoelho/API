## 3. Consumindo endpoints com Python

Vamos consultar os endpoints `/health` e `/info`.


````python
import requests

base_url = 'http://localhost:8000'

health = requests.get(f'{base_url}/health').json()
info = requests.get(f'{base_url}/info').json()

health, info
````

## 4. Validando comportamento esperado

Podemos validar as respostas esperadas com asserts simples.


````python
assert health == {'status': 'ok'}
assert info['name'] == 'API'
print('Respostas validadas com sucesso!')
````
